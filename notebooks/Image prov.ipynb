```python
from prov.dot import prov_to_dot
from prov.model import ProvDocument
from pathlib import Path

# Define o caminho para o seu arquivo PROV JSON
prov_json_file = Path('/content/pipeline_provenance.json')

# Verifica se o arquivo existe
if prov_json_file.exists():
    # Lê o conteúdo do arquivo JSON
    with open(prov_json_file, 'r', encoding='utf-8') as f:
        json_data = f.read()

    # Deserializa o JSON para um objeto ProvDocument
    d1 = ProvDocument.deserialize(content=json_data, format='json')

    # Converte o documento de proveniência para um objeto Graphviz Dot
    dot = prov_to_dot(d1)

    # Define o caminho de saída para a imagem PNG
    output_png_path = 'sample_data/pipeline_provenance.png'

    # Salva o gráfico como uma imagem PNG
    dot.write_png(output_png_path)
    print(f"Imagem de proveniência salva em: {output_png_path}")
else:
    print(f"Erro: O arquivo PROV JSON não foi encontrado em {prov_json_file}.")
    print("Por favor, execute a função `generate_complete_pipeline_provenance()` ou verifique o caminho do arquivo.")
```
